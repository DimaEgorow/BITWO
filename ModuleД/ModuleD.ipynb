```python
import pandas as pd
import matplotlib.pyplot as plt
# Загрузка данных из CSV файла
file_path = 'C:/Users/yegor/Desktop/BI/ModuleД/updated_file.csv'
df = pd.read_csv(file_path)

# Фильтрация данных по информатике
informatics_data = df[df['Конкурс'].str.contains('Информатика', na=False)]

# Анализ данных для определения полезных тем
# Создадим словарь для подсчета упоминаний навыков
skills_count = {}

for index, row in informatics_data.iterrows():
    skills = row['Необходимые знания'].split(', ')
    for skill in skills:
        if skill in skills_count:
            skills_count[skill] += 1
        else:
            skills_count[skill] = 1

# Преобразуем словарь в DataFrame и сортируем по количеству упоминаний
skills_df = pd.DataFrame(list(skills_count.items()), columns=['Навык', 'Количество упоминаний'])
skills_df = skills_df.sort_values(by='Количество упоминаний', ascending=False)

# Построение маршрутного листа
route = skills_df.reset_index(drop=True)

# Вывод результатов в виде списка
print("Оптимальная образовательная траектория по навыкам:")
print(route)

# Визуализация данных
plt.figure(figsize=(10, 6))
plt.barh(route['Навык'], route['Количество упоминаний'], color='skyblue')
plt.xlabel('Количество упоминаний')
plt.title('Образовательная траектория по навыкам')
plt.gca().invert_yaxis()  # Инвертируем ось Y для лучшего отображения
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Обоснование метрик
average_participants = informatics_data['Количество участников'].mean()
print(f"\nСреднее количество участников по информатике: {average_participants:.2f}")

# Вычислительная сложность
# Сложность анализа: O(n * m), где n - количество конкурсов, m - среднее количество навыков в каждом конкурсе
print("\nСложность алгоритма: O(n * m)")
```

Сложность анализа составляет O(n * m), где n — количество конкурсов, а m — среднее количество навыков, упомянутых в каждом конкурсе. Это связано с тем, что мы проходим по всем конкурсам и разбиваем навыки на отдельные элементы.

3. Асимптотика выбранного алгоритма
Асимптотическая сложность алгоритма, используемого для построения образовательной траектории, может быть определена следующим образом:

Фильтрация данных: O(n), где n — количество конкурсов.
Итерация по строкам: O(m), где m — количество строк в отфильтрованном DataFrame.
Разделение строк на навыки: O(k) для каждой строки, где k — среднее количество навыков в каждой строке.
